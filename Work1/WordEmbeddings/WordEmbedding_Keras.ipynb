```python
df_qqp = qqp_df.head(100)
df_qqp.to_csv("df_qqp")
```

```python
# encoded each word in a question...
dim_size = 100
encode = [one_hot(i, dim_size) for i in df_qqp['question1']]
print(encode)
```

```python
# vector size should be same
max_length = max([len(i) for i in encode])
vector_trimming = pad_sequences(encode, maxlen = max_length, padding = 'pre')
print(vector_trimming)
```

```python
# Model Building - Word Embedding
model_build = Sequential()
model_build.add(Embedding(dim_size, 10, input_length = max_length))
model_build.add(Flatten())
model_build.add(Dense(1, activation = 'sigmoid'))
```

```python
model_build.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
model_build.fit(vector_trimming, df_qqp['label'], epochs = 30)

loss_function, accuracy_metrics = model_build.evaluate(vector_trimming, df_qqp['label'], verbose = 0)
print("\n Accuracy:", accuracy_metrics*100, "%")
```

```python
# Prediction
user_sentences = ["How to play this game?", "Do you feel happy today?", 
                 "What is bibliography?", "What is an anode?"]

dim_size_pre = 100
encode_pre = [one_hot(i, dim_size_pre) for i in user_sentences]
print(encode_pre)
```

```python
max_length_pre = 33
vector_trimming_pre = pad_sequences(encode_pre, maxlen = max_length_pre, padding = 'pre')
print(vector_trimming_pre)
```

```python
prediction = model_build.predict(vector_trimming_pre) # --- Probability
#prediction_execute = np.argmax(prediction, axis = 1)
print(prediction)
#print(prediction_execution)
predictions = (model_build.predict(vector_trimming_pre) > 0.5).astype("int32")
print(predictions)
```
