```python
import datasets as ds
import pandas as pd
import numpy as np
from gensim.models import Word2Vec
from gensim.models import KeyedVectors
import warnings
import keras
import tensorflow
import os
import warnings
```

```python
ds.list_datasets()
```

```python
cola_dataset = ds.load_dataset('glue', 'cola', split = 'train')
```

```python
cola_dataset.shape
```

```python
cola_df = cola_dataset.to_pandas()
```

```python
df_cola = cola_df
df_cola.head(10)
```

```python
cola_sentences = df_cola['sentence']
cola_list = list(cola_sentences)
cola_list
```

```python
cola_split = [i.split() for i in cola_list]
cola_split
```

```python
warnings.filterwarnings('ignore')
cola_model_build = Word2Vec(cola_split, min_count = 1) # training Word2Vec model
```

```python
print(cola_model_build)
```

```python
cola_words = list(cola_model_build.wv.vocab) # summarizing all the words in one list => unique words
print(cola_words)
print(len(cola_words))
```
